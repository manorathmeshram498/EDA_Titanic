```python
df["Title"] = df["Name"].str.extract(' ([A-Za-z]+)\.')
```

```python
sns.barplot(y = 'Survived', x = 'Title', data = df)
```

```python
def replace_title(row):
  title = row['Title']
  relevalnt_titles = ['Mr', 'Mrs', 'Miss', 'Master']

  if title in relevalnt_titles:
    return title
  else:
    return 'others'
```

```python
df['Title'] = df.apply(replace_title, axis =1)
```

```python
df
```

```python
sns.barplot(y = 'Survived', x = 'Title', data = df)
```
